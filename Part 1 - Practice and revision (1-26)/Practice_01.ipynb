```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid_toy.csv")
df
```

```python
sdf = df
```

```python
df.info()
```

```python
age_s = sdf['age']
fever_s = sdf['fever']



new_age = pd.DataFrame(age_s)
new_fever = pd.DataFrame(fever_s)

numeric = new_age
numeric['fever'] = new_fever
numeric
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
Scaler.fit(numeric)
```

```python
new_num = Scaler.transform(numeric)
```

```python
new_num
```

```python
new_vals = pd.DataFrame(new_num, columns=['Age', 'Fever'])
```

```python
new_vals
```

```python
sdf['age'] = new_vals['Age']
sdf['fever'] = new_vals['Fever']
```
